```python
import os
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

def dividend_policy(stock_id: str) -> pd.DataFrame:
    # 設定 Chrome options
    chrome_options = Options()
    chrome_options.add_argument('--disable-blink-features=AutomationControlled')

    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

    try:
        url = f"https://goodinfo.tw/tw/StockDividendPolicy.asp?STOCK_ID={stock_id}"
        driver.get(url)
        time.sleep(3)
        
        # 取得股利政策表格 HTML
        div = driver.find_element(By.ID, "divDetail")
        html = div.get_attribute("outerHTML")
        
        # 轉成 DataFrame
        df = pd.read_html(html)[0]

        return df

    except Exception as e:
        print(f"發生錯誤: {e}")
        return None

    finally:
        driver.quit()

dividend_policy("2033")
```

```python
import os
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

def xbrl(stock_id: str, year_quarter: str = None) -> None:
    """
    stock_id      - 股票代号（如 '2330'）
    year_quarter  - 期别（如 '113Q1'），不填默认下载最新一笔
    """

    # 下载路径为当前目录下的 xbrl_data 文件夹
    output_dir = os.path.join(os.getcwd(), "xbrl_data")
    os.makedirs(output_dir, exist_ok=True)

    # 设定 selenium 的下载目录
    chrome_options = Options()
    chrome_options.add_argument("--disable-blink-features=AutomationControlled")
    chrome_options.add_experimental_option("prefs", {
        "download.default_directory": output_dir,
        "download.prompt_for_download": False,
        "download.directory_upgrade": True
    })

    # 启动浏览器
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

    try:
        print("✅ 開啟 MOPS 觀測站...")
        driver.get("https://mops.twse.com.tw/mops/web/t203sb01")
        time.sleep(7)

        # 输入股票代号
        print(f"✅ 輸入股票代號 {stock_id}")
        input_box = driver.find_element(By.NAME, "co_id")
        input_box.clear()
        input_box.send_keys(stock_id)
        time.sleep(0.5)

        # 执行查询按钮（onclick 为 js 函数）
        print("✅ 點擊查詢")
        search_button = driver.find_element(By.XPATH, '//input[@type="button" and @value="查詢"]')
        search_button.click()
        time.sleep(3)

        # 表格加载
        print("✅ 等待結果表格載入")
        table = driver.find_element(By.ID, "table01")
        rows = table.find_elements(By.TAG_NAME, "tr")

        # 选择目标期别
        target_row = None
        for row in rows[1:]:
            cols = row.find_elements(By.TAG_NAME, "td")
            if not cols:
                continue
            period = cols[0].text.strip()
            if year_quarter is None or period == year_quarter:
                target_row = row
                print(f"✅ 選定期別：{period}")
                break

        if not target_row:
            print("❌ 找不到符合的年季別")
            return

        # 点击下载按钮
        print("✅ 點擊下載連結")
        download_link = target_row.find_element(By.PARTIAL_LINK_TEXT, "下載")
        download_link.click()

        # 等待下载完成
        time.sleep(10)
        print(f"✅ 檔案已下載至：{output_dir}")

    except Exception as e:
        print(f"❌ 發生錯誤：{e}")
    finally:
        driver.quit()
```
